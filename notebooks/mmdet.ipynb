```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  
```

```python
from decord import VideoReader
from decord import cpu, gpu

vr = VideoReader('../data/videos/ydcrodwtz3mstjq1vhbdflx6kyhj3y0p.mp4', ctx=cpu(0))
```

```python
image = vr[1600*10].asnumpy()
```

```python
plt.imshow(image)
```

```python
from PIL import Image
```

```python
Image.fromarray(image).save("example.png")
```

```python
results = model(image)
```

```python
import cv2
import matplotlib.pyplot as plt
for result in results:
    boxes = result.boxes  # get bounding box coordinates
    for i, box in enumerate(boxes[7:9]):
        print(i)
        x1, y1, x2, y2 = map(int, box.xyxy[0, :4])  # convert to integers
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # draw rectangle on image
        plt.imshow(image)  # display the image
        plt.show()
plt.imshow(image)  # display the image
plt.show()
```
